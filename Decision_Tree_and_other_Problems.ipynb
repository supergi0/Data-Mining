```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
x = df1.drop(1 , axis = 1)
y=df1[1]
y2=[]
for tumor in y:
  if(tumor=='M'):
    y2.append(1)
  else:
    y2.append(0)
y2=np.array(y2)
y2=y2.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y2, test_size=0.2, random_state=0)
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
class Node():
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, info_gain=None, value=None):
        self.feature_index = feature_index
        self.threshold = threshold
        self.left = left
        self.right = right
        self.info_gain = info_gain
        self.value = value
```

```python
class DecisionTree():
    def __init__(self, min_samples_split=5, max_depth=10):
        self.root = None
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
  
    def split(self, dataset, feature_index, threshold):
        left = np.array([row for row in dataset if row[feature_index]<=threshold])
        right = np.array([row for row in dataset if row[feature_index]>threshold])
        return left, right

    def entropy(self, y):
        labels = np.unique(y)
        ent = 0
        for cls in labels:
            p_cls = len(y[y == cls]) / len(y)
            ent += -p_cls * np.log2(p_cls)
        return ent
    
    def gini_index(self, y):
        labels = np.unique(y)
        gini = 0
        for cls in labels:
            p_cls = len(y[y == cls]) / len(y)
            gini += p_cls**2
        return 1 - gini    

    def info_gain(self, parent, l_child, r_child, mode="entropy"):
        wl = len(l_child) / len(parent)
        wr = len(r_child) / len(parent)
        if mode=="gini":
            gain = self.gini_index(parent) - (wl*self.gini_index(l_child) + wr*self.gini_index(r_child))
        else:
            gain = self.entropy(parent) - (wl*self.entropy(l_child) + wr*self.entropy(r_child))
        return gain

    def get_best_split(self, dataset,samples, features):
        best_split = {}
        max_info_gain = -float("inf")
        for feature_index in range(features):
            feature_values = dataset[:, feature_index]
            possible_thresholds = np.unique(feature_values)
            for threshold in possible_thresholds:
                left, right = self.split(dataset, feature_index, threshold)
                if len(left)>0 and len(right)>0:
                    y, left_y, right_y = dataset[:, -1], left[:, -1], right[:, -1]
                    info_gain = self.info_gain(y, left_y, right_y, "gini")
                    if info_gain>max_info_gain:
                        best_split["feature_index"] = feature_index
                        best_split["threshold"] = threshold
                        best_split["dataset_left"] = left
                        best_split["dataset_right"] = right
                        best_split["info_gain"] = info_gain
                        max_info_gain = info_gain
        return best_split

    def build_tree(self, dataset, depth=0):
        X, Y = dataset[:,:-1], dataset[:,-1]
        samples, features = np.shape(X)
        if samples >= self.min_samples_split and depth <= self.max_depth:
            best_split = self.get_best_split(dataset, samples, features)
            if best_split["info_gain"] > 0:
                left_subtree = self.build_tree(best_split["dataset_left"], depth+1)
                right_subtree = self.build_tree(best_split["dataset_right"], depth+1)
                return Node(best_split["feature_index"], best_split["threshold"], 
                            left_subtree, right_subtree, best_split["info_gain"])
        leaf_value = self.calculate_leaf_value(Y)
        return Node(value=leaf_value)
        
    def calculate_leaf_value(self, Y):
        Y = list(Y)
        return max(Y, key=Y.count)
    
    def fit(self, X, Y):
        dataset = np.concatenate((X, Y), axis=1)
        self.root = self.build_tree(dataset)

    def prediction(self, x, tree):
        if tree.value!=None: return tree.value
        feature_val = x[tree.feature_index]
        if feature_val<=tree.threshold:
            return self.prediction(x, tree.left)
        else:
            return self.prediction(x, tree.right)

    def predict(self, X):
        preditions = [self.prediction(x, self.root) for x in X]
        return preditions
```

```python
classifier = DecisionTree()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test) 
accuracy_score(y_test, y_pred)
```
